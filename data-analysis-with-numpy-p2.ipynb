###  ndarray Manipulation -1

```python
import numpy as np

arr = np.random.randint(1, 100, size=(10, 5))  #sample-10, feature-5
```

```python
arr.shape
```

```python
# reshape of array

b = arr.reshape(5, 10)

b.shape
b.ndim
b
```

```python
# dimention convert 

flatten = b.flatten()
#row wise combine
print(flatten)
print(flatten.ndim)

#column wise combine

columnWiseFlattening = np.ravel(b, order= 'F')

print(columnWiseFlattening)
```

## arithmetic operator and mathematical function

```python
x = np.array([10, 20, 30, 40])
y = np.array([2, 3, 4, 5])

add = x + y
print(add)

sub = x - y
print(sub)

mul = x * y
print(mul)

div = x / y
print(div)

remainder = x % y
print(remainder)
```

```python
# mathematical function

#trigonometry

#sin, cos, tan takes values in radian

sin_val = np.sin(x)
print(sin_val)

cos_val = np.cos(x)
print(cos_val)

deg_con = np.rad2deg(x)
print(deg_con)

rad_con = np.deg2rad(deg_con)
print(rad_con)

```

```python
#(google search numpy mathematical functions)

#log
x = np.array([10, 20, 30, 40, 100, 16])
base_10_log_val = np.log10(x)
print(base_10_log_val)

div_by_2 = np.log2(x)
print(div_by_2)

sqrt = np.sqrt(x)
print(sqrt)

sum = np.sum(x)
print(sum)

cumul = np.cumulative_sum(x)
print(cumul)
```

## sorting

```python
# inplace 
x = np.array([10, 8, 16, 100])

z= x.copy()
print(z)

z.sort()

print(z)
```

```python
#copy sorting
print(x)

sort_arr = np.sort(x)

print(sort_arr)

print(x)
```

```python
#2d array sorting

mat = np.array([
    [10, 3, 5],
    [8, 4, 9]
])

# horizon axis = 1

hor_sort = np.sort(mat, axis= 1)
print(hor_sort)

# vartical axis = 0

vart_sort = np.sort(mat, axis= 0)
print(vart_sort)
```

## searching

```python
print(x)
```

```python
# np.where(condition, x, y) if true -> x, otherwise replace with y 

index = np.where(x == 16) #  returns index

print(index)

arr = np.where(x > 8, x, 0) # returns array

print(arr)
```

```python
x = np.array([10,100, 8, 16, 100])

max_val_indx = np.argmax(x)
print(max_val_indx)

min_val_indx = np.argmin(x)
print(min_val_indx)
```

## counting in ndarray

```python
a = np.random.randint(1, 100, size=(100,))
print(a)
```

```python
val_greater_than_60 = np.count_nonzero(a == 21)
print(val_greater_than_60)
```

```python
unique_val, count = np.unique(a, return_counts= True)
print(unique_val)
print(count)
```
