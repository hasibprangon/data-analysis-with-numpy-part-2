###  ndarray Manipulation -1

```python
import numpy as np

arr = np.random.randint(1, 100, size=(10, 5))  #sample-10, feature-5
```

```python
arr.shape
```

```python
# reshape of array

b = arr.reshape(5, 10)

b.shape
b.ndim
b
```

```python
# dimention convert 

flatten = b.flatten()
#row wise combine
print(flatten)
print(flatten.ndim)

#column wise combine

columnWiseFlattening = np.ravel(b, order= 'F')

print(columnWiseFlattening)
```

```python
a = np.random.randint(1, 10, size=(2, 3))  #sample-10, feature-5

b = np.random.randint(20, 30, size=(2, 3))

print(a)
print(b)
```

### concatenate

```python
con_row = np.concatenate((a,b), axis= 0) # row wise

print(con_row)

con_col = np.concatenate((a,b), axis= 1) # col wise
print(con_col)
```

## arithmetic operator and mathematical function

```python
x = np.array([10, 20, 30, 40])
y = np.array([2, 3, 4, 5])

add = x + y
print(add)

sub = x - y
print(sub)

mul = x * y
print(mul)

div = x / y
print(div)

remainder = x % y
print(remainder)
```

```python
# mathematical function

#trigonometry

#sin, cos, tan takes values in radian

sin_val = np.sin(x)
print(sin_val)

cos_val = np.cos(x)
print(cos_val)

deg_con = np.rad2deg(x)
print(deg_con)

rad_con = np.deg2rad(deg_con)
print(rad_con)

```

```python
#(google search numpy mathematical functions)

#log
x = np.array([10, 20, 30, 40, 100, 16])
base_10_log_val = np.log10(x)
print(base_10_log_val)

div_by_2 = np.log2(x)
print(div_by_2)

sqrt = np.sqrt(x)
print(sqrt)

sum = np.sum(x)
print(sum)

cumul = np.cumulative_sum(x)
print(cumul)
```
